```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fruit = ["Apple", "Orange", "Grapes", "Banana"]
quantity = [67, 34, 100, 29]
```

```python
plt.pie(quantity, labels=fruit)
plt.show()
```

```python
plt.pie(quantity, labels=fruit, autopct="%0.1f%%", colors=["pink", "yellow", "magenta", 'cyan'])
plt.show()
```

```python
# Doughnut chart
```

```python
plt.pie(quantity, labels=fruit, radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.show()
```
