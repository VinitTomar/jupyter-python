```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = [1,   4,  3,  5, 99, 10, 44]
y = [55, 66, 88, 23,  4, 87, 73]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, marker="*", color="g", s=100)
plt.show()
```

```python
y2 = [33, 54, 23, 76, 88, 9, 10]
```

```python
plt.scatter(x, y, marker="*", color="g", s=100)
plt.scatter(x, y2, marker=".", color="r", s=100)
plt.show()
```
