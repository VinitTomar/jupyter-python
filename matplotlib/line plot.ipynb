```python
# Line plot
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 11)
x
```

```python
y = x * 2
y
```

```python
#plt.plot(x, y)
plt.plot(x, y, color='g', linestyle=":", linewidth=2)
plt.title("Simple plot")
plt.xlabel("X vals")
plt.ylabel("Y vals")
plt.show()
```

```python
y2 = x * 3
y2
```

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.grid(True)
plt.show()
```

```python
# sub plots
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(x, y, color="r")
plt.grid(True)

plt.show()
```
